#### DataFrame Index /  Columns
- index : 인덱스 따로 지정
- columns : 데이터프레임 열 지정
- OrderedDict : 순서가 보장된 딕셔너리 전달

```python
scientists = pd.DataFrame( data = {
    'Name':['Rosaline Franklin','William Gosset'],
    'Occupation':['Chemist','Statistician'],
    'Born':['1920-07-25','1876-06-13'],
    'Died':['1958-04-16','1937-10-16'],
    'Age':[37,61] },
    index = ['Rosaline Franklin','William Gosset'],
    columns = ['Occupation','Born','Age','Died'])
print(scientists)
```

```python
from collections import OrderedDict
scientists = pd.DataFrame(OrderedDict([
    ('Name',['Rosaline Franklin','William Gosset']),
    ('Occupation',['Chemist','Statistician']),
    ('Born',['1920-07-25','1876-06-13']),
    ('Died',['1958-04-16','1937-10-16']),
    ('Age',[37,61])
]))                   
print(scientists)
```

## 03-3. 시리즈 다루기-응용

#### boolean 추출
- 리스트의 형태로 참 / 거짓을 담아 시리즈에 전달하면 참인 인덱스의 데이터만 추출한다.

```python
scientists = pd.read_csv('C:\\data\\scientists.csv')
```

```python
print(scientists.shape)
```

```python
print(scientists.head(10))
```

```python
ages = scientists['Age']
print(ages.max())
```

```python
print(ages.mean())
```

```python
print(ages>ages.mean())
```

```python
print(type(ages>ages.mean()))
```

```python
print(ages[ages>ages.mean()])
```

#### Series와 Broadcasting
- 브로드캐스팅 : 시리즈나 데이터프레임에 있는 모든 데이터에 대해 한 번에 연산하는 것
- 일치하는 인덱스끼리 브로드캐스팅 수행
- 길이가 서로 다른 벡터의 연산 => 일치하는 인덱스만 수행하고 나머지는 누락값(NaN)으로 처리 

```python
print(ages+ages)
```

```python
print(ages*ages)
```

```python
print(ages*2)
```

```python
print(pd.Series([1,100]))
```

```python
print(ages+pd.Series([1,100]))
```

```python
print(ages)
```

```python
rev_ages = ages.sort_index(ascending=False)
print(rev_ages)
```

```python
print(ages+rev_ages)
```

```python
print(ages*2)
```

```python
print(scientists*2)   #문자열은 2배 길어진다.
```

## 03-5. 시리즈와 데이터프레임의 데이터 처리하기

#### 열 자료형 바꾸기 
- 날짜형 데이터를 문자형이 아닌 datetime으로 바꾸기

```python
print(scientists)
```

```python
print(scientists['Born']) # dtype : object
```

```python
born_datetime = pd.to_datetime(scientists['Born'],format='%Y-%m-%d')
print(born_datetime) # dtype : datetime 
```

```python
print(scientists['Died']) # dtype : object
```

```python
died_datetime = pd.to_datetime(scientists['Died'],format='%Y-%m-%d')
print(died_datetime) # dtype : datetime 
```

#### 데이터프레임에 새로운 열 추가하기

```python
scientists['born_dt'],scientists['died_dt']=(born_datetime, died_datetime)
```

```python
print(scientists)
```

```python
scientists['age_days_dt']=(scientists['died_dt']-scientists['born_dt'])
```

```python
print(scientists)
```
